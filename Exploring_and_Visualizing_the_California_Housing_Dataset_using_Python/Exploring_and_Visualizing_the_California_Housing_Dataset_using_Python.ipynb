```python
numerical_features = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']
df[numerical_features].hist(bins=30, figsize=(15, 10), layout=(3, 3))
plt.tight_layout()
plt.show()
```

# **Geographical**  **Distribution of House Values**

Visualize the geographical distribution of house values using latitude and longitude.


```python
plt.scatter(df['longitude'], df['latitude'], c=df['median_house_value'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Median House Value')
plt.title('Geographical Distribution of House Values')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
sns.boxplot(x=df['median_house_value'])
plt.title('Box Plot of Median House Value')
plt.xlabel('Median House Value')
plt.show()
```

#**Pair Plot**

```python
sns.pairplot(df[['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']])
plt.show()
```

#**Box Plot of Median House Value**
Identify outliers in the distribution of median_house_value.

```python
sns.boxplot(x=df['median_house_value'])
plt.title('Box Plot of Median House Value')
plt.xlabel('Median House Value')
plt.show()
```

# Correlation Heatmap
Display the correlation between different numerical features to understand their relationships.

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

#Median Income vs. Median House Value
Visualize the relationship between median_income and median_house_value.

```python
plt.scatter(df['median_income'], df['median_house_value'])
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
sns.pairplot(df[['median_income', 'median_house_value', 'housing_median_age']])
plt.show()
```
